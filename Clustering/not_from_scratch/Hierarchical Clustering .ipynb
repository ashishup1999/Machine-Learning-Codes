```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Mall_Customers.csv")
print(df[:5])
```

```python
x=df.iloc[:,[3,4]].values
print(x[:5])
```

```python
# onehot encoding
'''from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
x = np.array(ct.fit_transform(x))'''
```

```python
k=5 #no of clusters.
```

```python
#dendogram for getting no of clusters requred for optimal result
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
#from above dendogram we can know what number of cluster is required.
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=k)
classes=cluster.fit_predict(x)
```

```python
# Visualising the Training set results
cls_clr={0:"red",1:"green",2:"blue",3:"cyan",4:"yellow"}
for i in range(x.shape[0]):
    plt.scatter(x[i:,0],x[i:,1], color=cls_clr[classes[i]])
plt.title('Clustering')
plt.xlabel('Age')
plt.ylabel('Anual Income')
plt.show()
```
